```python
import pandas as pd

hero_attribute = './Datasets/superhero-set/heroes_included_information.csv'
movie_data = './Datasets/movie_imdb_ratings_CC.csv'
movie_superhero_list = './Datasets/superhero_movies_data.csv'

attribute_df = pd.read_csv(hero_attribute)
movie_rating_sales_df = pd.read_csv(movie_data)
movie_superhero_df = pd.read_csv(movie_superhero_list)
```

```python
movie_superhero_df.head()
```

```python
movie_rating_sales_df.rename(columns={'Internet_Movie_Database_Rating':'Rating'},inplace=True)
movie_rating_sales_df.head()
```

```python
del movie_rating_sales_df['Unnamed: 0']
```

```python
movie_rating_sales_df.head()
```

```python
merge1 = pd.merge(movie_superhero_df, movie_rating_sales_df, on = 'Movie_Title', how= 'outer')
merge1.head()
```

```python
attribute_df.head()
```

```python
attribute_df.rename(columns={'name':'Superhero_Name'},inplace=True)
attribute_df.head()
```

```python
merge = pd.merge(merge1, attribute_df, on = 'Superhero_Name', how= 'left')
merge
```

```python
del merge['Date_Released_y']
del merge['Publisher_y']
del merge['Unnamed: 0']
merge.rename(columns={'Date_Released_x':'Date_Released','Publisher_x':'Publisher'}, inplace= True)
merge.head()
```

```python
Inflation_df = pd.read_csv('./Datasets/Inflation_adjustments.csv') 
Inflation_df.head()
```

```python
merge['Box_Office_Sales'] = merge['Box_Office_Sales'].str.replace('$','')
merge['Box_Office_Sales'] = merge['Box_Office_Sales'].str.replace(',','')
merge['Box_Office_Sales'] = merge['Box_Office_Sales'].str.replace(' ','')
merge['Box_Office_Sales'] = merge['Box_Office_Sales'].astype(int)
merge.head()
```

```python
merge['Rating'] = merge['Rating'].str.replace('/10','')
merge['Rating'] = merge['Rating'].str.replace(' ','')
merge['Rating'] = merge['Rating'].astype(float)
merge['Rating'] = merge['Rating'] * 10
merge.head()
```

```python
rate = []
for index , value in enumerate(merge['Date_Released']):
    number = Inflation_df[Inflation_df['Year'] == value]['Rate'].values[0]
    #print(number)
    rate.append(number)
merge['rate'] = rate
merge.head()
```

```python
merge['Inflation_Sales'] = ((merge['rate'] /100) + 1)* merge['Box_Office_Sales']
merge.head()
```

```python
# save file
merge.to_csv('./Datasets/Merged_Superhero.csv')
```
