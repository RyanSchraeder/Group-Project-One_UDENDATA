```python
# import dependencies
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# movie data query from omdbapi.com

query_url = "http://www.omdbapi.com/?apikey=cf77db42&t="
print(query_url)
```

```python
movie_df = pd.read_csv('./Datasets/superhero_movies_data.csv')
movie_df.head()
```

```python
# selected movies
# movies = ['Ant-Man',
#           'Ant-Man and the Wasp',
#           'Aquaman',
#           'The Dark Knight',
#           'The Dark Knight Rises',
#           'Black Panther',
#           'Captain America: The First Avenger',
#           'Captain America: The Winter Soldier',
#           'Captain Marvel',
#           'Deadpool',
#           'Deadpool 2',
#           'Doctor Strange',
#           'Spirit of Vengeance',
#           'Green Lantern',
#           'Iron Man',
#           'Iron Man 2',
#           'Iron Man 3',
#           'Punisher: War Zone',
#           'Spider-Man: Homecoming',
#           'The Amazing Spider-Man',
#           'The Amazing Spider-Man 2',
#           'Man of Steel',
#           'The Incredible Hulk',
#           'Thor',
#           'Thor: Ragnarok',
#           'Thor: The Dark World',
#           'Venom',
#           'Logan',
#           'The Wolverine',
#           'X-Men Origins: Wolverine',
#           'Wonder Woman',
# ]

# variables
year = []
title = []
ratings = []
imdbVotes = []
response = []
sales = []

movie_data_accumulated = []

# loop through database
for movie in movie_df['Movie_Title']:
             
    response = requests.get(query_url + movie).json()
    year = response['Year']
    title = response['Title']
    ratings = response['Ratings'][0]['Value']
    imdbVotes = response['imdbVotes']
    sales = response['BoxOffice']
    
    #print(year, title, ratings)
    movie_data = [year, title, ratings, imdbVotes, sales]
    #print(movie_data)
    movie_data_accumulated.append(movie_data)

# create dataframe

movie_data_df = pd.DataFrame(movie_data_accumulated, columns=('Date_Released', 
                                                              'Movie_Title',
                                                              'Internet_Movie_Database_Rating',
                                                              'IMDB_Votes', 
                                                              'Box_Office_Sales'))
movie_data_df
```

```python
# save file
movie_data_df.to_csv('./Datasets/movie_imdb_ratings_CC.csv')
```
