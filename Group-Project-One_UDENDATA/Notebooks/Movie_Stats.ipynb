```python
# import dependencies
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# movie data query from omdbapi.com

query_url = "http://www.omdbapi.com/?apikey=cf77db42&t="
print(query_url)
```

```python
movie_df = pd.read_csv('./Datasets/Production_data/Movie_List.csv')
movie_df.head()
```

```python
# variables
year = []
#title = []
ratings = []
imdbVotes = []
response = []
sales = []

title = movie_df['Movie_Title']
movie_data_accumulated = []

# loop through database
for movie in movie_df['Movie_Title']:
             
    response = requests.get(query_url + movie).json()
    year = response['Year']
    #title = response['Title']
    ratings = response['Ratings'][0]['Value']
    imdbVotes = response['imdbVotes']
    sales = response['BoxOffice']
    
    #print(year, title, ratings)
    movie_data = [year, ratings, imdbVotes, sales]
    #print(movie_data)
    movie_data_accumulated.append(movie_data)

# create dataframe
movie_data_df = pd.DataFrame(movie_data_accumulated, columns=('Date_Released', 
                                                              'Internet_Movie_Database_Rating',
                                                              'IMDB_Votes', 
                                                              'Box_Office_Sales'))
movie_data_df['Movie_Title'] = title
movie_data_df
```
