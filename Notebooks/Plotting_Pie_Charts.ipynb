## Pie Charts for Category

- Superhero alignment percentage
- Superhero human vs. non-human percentage

## Bar Charts for Sales 

- Sales by Alignment percentage
- Sales by Human Vs. Non-Human percentage 

## Bar Charts for Ratings

- Ratings by Alignment percentage
- Ratings by Human Vs. Non-Human percentage

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
superhero_data = pd.read_csv('../Datasets/Production_data/Merged_Superhero.csv')
superheroes_unique = superhero_data.drop_duplicates(subset=['Superhero_Name'], keep='first') 
superheroes_unique
```

```python
superhero_alignment = superheroes_unique['Alignment'].value_counts()
superhero_alignment_df = pd.DataFrame(superhero_alignment)
superhero_alignment_df
```

### Pie Chart - Alignment Percentage

```python
labels = ['Good', 'Neutral', 'Bad']
colors = ['lightskyblue', 'lightcoral', 'red']
explode = [0.0, 0.0, 0.1]
plt.title('Good vs. Bad')
plt.pie(superhero_alignment_df, labels=labels, colors=colors, explode=explode, shadow=True, autopct="%1.2f%%")
plt.show()
```

### Pie Chart - Alignment Percentage - Good Vs. Bad

```python
superheroes_bad = superhero_alignment_df.loc[superhero_alignment_df.index != 'good']
superheroes_bad
```

### Pie Chart - Race Percentage

```python
superhero_race_count = superheroes_unique['Race'].value_counts()
superhero_race_df = pd.DataFrame(superhero_race_count)
superhero_race_df
```

```python
labels = ['Human', 'Non-Human']
colors = ['lightskyblue', 'green']
explode = [0.0, 0.1]
plt.title('Human vs. Non-Human')
plt.pie(superhero_race_df, labels=labels, colors=colors, explode=explode,
       shadow=True)
plt.show()
```

### Bar Chart - Sales by Alignment

```python
# bar chart for sales by alignment    
sales_grouped = superheroes_unique.groupby(['Alignment'])

sales_by_alignment = sales_grouped['Box_Office_Sales'].sum()
sales_by_alignment_df = pd.DataFrame({'Sales': sales_by_alignment})
sales_by_alignment_df

```

```python
# plot bar chart for sales by alignment

x_axis = np.arange(len(sales_by_alignment_df))
plt.xticks(x_axis,sales_by_alignment_df.index, rotation=45)
plt.bar(x_axis,sales_by_alignment_df['Sales'], alpha=.5, align='center', color='blue')
plt.title('Good vs. Bad Sales')
plt.ylabel('Total Sales(Hundred Millions)')
plt.show()
```

### Bar Chart - Sales by Race

```python
# bar chart for sales by race    
sales_grouped_race = superheroes_unique.groupby(['Race'])
sales_grouped_race
sales_by_race = sales_grouped_race['Box_Office_Sales'].sum()
sales_by_race
sales_by_race_df = pd.DataFrame({'Sales': sales_by_race})
sales_by_race_df
```

```python
# plot bar chart for sales by race

x_axis = np.arange(len(sales_by_race_df))
plt.xticks(x_axis,sales_by_race_df.index, rotation=45)
plt.bar(x_axis,sales_by_race_df['Sales'], alpha=.5, align='center', color='blue')
plt.title('Human vs. Non-Human Sales')
plt.ylabel('Total Sales(Hundred Millions)')
plt.show()
```

### Bar Chart - Ratings by Alignment

```python
# bar chart for ratings by alignment    
ratings_grouped = superheroes_unique.groupby(['Alignment'])

ratings_by_alignment = ratings_grouped['Rating'].mean()
ratings_by_alignment_df = pd.DataFrame({'Average Rating': ratings_by_alignment})
ratings_by_alignment_df
```

```python
# plot bar chart for ratings by alignment

x_axis = np.arange(len(ratings_by_alignment_df))
plt.xticks(x_axis,ratings_by_alignment_df.index, rotation=45)
plt.bar(x_axis,ratings_by_alignment_df['Average Rating'], alpha=.5, align='center', color='blue')
plt.title('Ratings by Good vs. Bad')
plt.ylabel('Total Sales(Hundred Millions)')
plt.show()
```

### Bar Chart - Sales by Race

```python
# # bar chart for sales by race   
sales_grouped_race = superheroes_unique.groupby(['Race'])

sales_by_race = sales_grouped_race['Box_Office_Sales'].sum()
sales_by_race_df = pd.DataFrame({'Sales': sales_by_race})
sales_by_race_df
```
